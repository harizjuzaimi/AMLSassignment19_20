```python
import A2_landmarks as l2
import numpy as np 
from sklearn.model_selection import train_test_split

def get_data():

    X_A2, y = l2.extract_features_labels()
    Y_A2 = np.array([y, -(y - 1)]).T
    

    return X_A2, Y_A2
```

```python
X_A2, Y_A2 = get_data()
```

```python
print(len(X_A2))
print(len(Y_A2))
```

```python
print(X_A2)
print(Y_A2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
temp = X_A2.reshape(len(X_A2), 68*2)
temp_X_A2 = scaler.fit_transform(temp)
temp_Y_A2 = list(zip(*Y_A2))[0]
```

```python
print(temp_X_A2)
print(temp_Y_A2)
```

```python
import pickle

with open('X_A2_saved.dat', 'wb') as f:
    pickle.dump(temp_X_A2, f) 
    
with open('Y_A2_saved.dat', 'wb') as f:
    pickle.dump(temp_Y_A2, f) 
```

```python
with open('X_A2_saved.dat', 'rb') as f:
    X_A2_load = pickle.load(f)
    
with open('Y_A2_saved.dat', 'rb') as f:
    Y_A2_load = pickle.load(f)
```
