```python
import B1_landmarks as b1l
import numpy as np 

def get_data():

    X_B1, y, _ = b1l.extract_features_labels()
    Y_B1 = np.array([y, -(y - 1)]).T
    

    return X_B1, Y_B1
```

```python
X_B1, Y_B1 = get_data()
```

```python
print(len(X_B1))
print(len(Y_B1))
```

```python
print(X_B1)
print(Y_B1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
temp = X_B1.reshape(len(X_B1), 68*2)
temp_X_B1 = scaler.fit_transform(temp)
temp_Y_B1 = list(zip(*Y_B1))[0]
```

```python
print(temp_X_B1)
print(temp_Y_B1)
```

```python
import pickle

with open('X_B1_saved.dat', 'wb') as f:
    pickle.dump(temp_X_B1, f) 
    
with open('Y_B1_saved.dat', 'wb') as f:
    pickle.dump(temp_Y_B1, f) 
```

```python
with open('X_B1_saved.dat', 'rb') as f:
    X_B1_load = pickle.load(f)
    
with open('Y_B1_saved.dat', 'rb') as f:
    Y_B1_load = pickle.load(f)
```
